```python
#reset_index.fillna(0)----replace nan and fill with 0
basket = (
    df.loc[df['Country'] == 'France']
    .groupby(['InvoiceNo', 'Description'])['Quantity']
    .sum()
    .unstack()
    .reset_index()
    .fillna(0)
    .set_index('InvoiceNo')
)
basket
```

```python
def encode(x):
  if x>0:
    return 1
  if x<=0:
    return 0

encoded_basket=basket.applymap(encode)

#applymap changes each datapoint value to 1 or 0 according to the condition
```

```python
encoded_basket
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
apr=apriori(encoded_basket,min_support=0.07,use_colnames=True)
```

```python
rules=association_rules(apr)
rules=rules.sort_values(['confidence'],ascending=False)
rules

# antecedents (A)

# consequents (B)
#antecedent support S(A)/N

#consequent support S(B)/N

#At a time we can calculate only two products only


```

```python
def get_recommendations(product,rules,num_recommendations=5):
  recommendations=[]
  for idk,row in rules.iterrows():
    if product in row['antecedents']:
      recommended_items=next(iter(row['consequents']))
      recommendations.append((recommended_items,row['lift']))
  recommendations.sort(key=lambda x:x[1],reverse=True)
  return recommendations[:num_recommendations]
```
